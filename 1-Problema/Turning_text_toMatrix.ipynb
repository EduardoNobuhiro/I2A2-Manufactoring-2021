```python
def remove_punct(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
text['Texto'] = text.Texto.map(remove_punct)
text
```

```python
def getuniquewords(dataset,n):
  coluna = dataset.iloc[:, n] 
  list_words = []
  count = 0
  while count < len(dataset.index):
    list_words = list_words+coluna[count].split()
    count = count+1
  list_words=list(set(list_words))
  return list_words
```

```python
final_list=getuniquewords(text,0)
final_list
```

```python
def text_toMatrix (dataset,n):
  columns1=getuniquewords(dataset,n)
  column_name=dataset.columns[n]
  dataframe = pd.DataFrame(columns=columns1)
  line = [0]*len(columns1)
  count1 = 0
  while count1 < len(dataset.index):
    count = 0
    while count < len(columns1):
      word_counted = dataframe.columns[count]
      line[count] = dataset[column_name][count1].count(word_counted)
      count += 1
    count_row = pd.Series(line, index = dataframe.columns)
    dataframe = dataframe.append(count_row, ignore_index=True)
    count1 += 1
  return dataframe
```

```python
matrix = text_toMatrix(text,0)
matrix
```
